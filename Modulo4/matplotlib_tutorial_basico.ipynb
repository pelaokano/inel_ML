```python
y1=np.random.randn(100).cumsum()
y2=np.random.randn(100).cumsum()
y3=np.random.randn(100).cumsum()
y4=np.random.randn(100).cumsum()
```

```python
fig=plt.figure(figsize=(20,20))
fig.add_subplot(2,2,1)
plt.plot(y1,"r-.")
fig.add_subplot(2,2,2)
plt.plot(y2,"g--")
fig.add_subplot(2,2,3)
plt.plot(y3,"b:")
fig.add_subplot(2,2,4)
plt.plot(y4,"y-")
plt.show()
```

```python
#se instancia el objeto figura y ejes
#se define el parametro linewidth, ancho de la linea del borde de la figura
fig,ax=plt.subplots(2,2,linewidth=12)
#la siguiente sentencia permite definir el tamao de la figura o el grafico
#ancho y alto
fig.set_size_inches(20,20)
#se define el color de la figura, sirve como fondo para los ejes
fig.set_facecolor("g")
#se define el color de la linea de borde de la figura
fig.set_edgecolor("b")
#permite agregar un titulo al grafico completo
plt.suptitle("Titulo grafico",fontsize=18)

#se ingresa valores sobre el eje
#ax[0,0] son una referencia a los ejes x e y
#g1 hace referencia a las curvas dentro de los ejes
#g1[0] es la primera curva dentro de los ejes que se indican antes
#para hacer referencia a las label y titulo de los ejes
#se debe hacer referencia a los ejes en este caso ax[0,0]
#ejemplo:
#title=ax[0,0].set_title("titulo 2")
#ax[0,1].set_xlabel("label x")
#ax[0,1].set_ylabel("label y")
#para hacer referencia a grosor de la curva y si tipo de 
#linea se debe hacer referencia a la curva, ejemplo
#g2[0].set_linewidth(5)
#g2[0].set_linestyle("-.")

g1=ax[0,0].plot(y1)
#ingresar el titulo de los subgraficos
title=ax[0,0].set_title("titulo 2")
#se ajusta el tamao de la letra
title.set_fontsize(18)
#se ajusta el color del titulo
title.set_color("r")
ax[0,1].set_xlabel("label x")
ax[0,1].set_ylabel("label y")
#se ajusta los limites de los ejes en x e y
ax[0,1].set_xlim(20,60)
ax[0,1].set_ylim(0,20)
#se ajusta el aspecto o relacion entre el eje x e y
ax[0,1].set_aspect("auto")
#se ingresan valores sobre otro eje
g2=ax[0,1].plot(y2)
#se ajusta el grosor de la linea
g2[0].set_linewidth(5)
g2[0].set_linestyle("-.")
#se ingresan valores sobre otro eje
g3=ax[1,0].plot(y3)
#se ingresan valores sobre otro eje
g4=ax[1,1].plot(y4)
plt.show()
```
